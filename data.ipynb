```python
import csv
from tqdm import tqdm

# Load existing data if the CSV already has entries
try:
    with open('responses.csv', 'r', newline='', encoding='utf-8') as csvfile:
        reader = csv.reader(csvfile)
        response_l = [row[0] for row in reader if row]  # Adjust to skip header if needed
        start_index = len(response_l)  # Start from where it left off
except FileNotFoundError:
    response_l = []
    start_index = 1  # Start from the beginning

# Write header if starting a new file
if start_index == 1:
    with open('responses.csv', 'w', newline='', encoding='utf-8') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(["response"])  # Write header only once

# Loop to generate and append responses
for i in tqdm(range(start_index, 18830)):
    prompt = f"""You are an expert sports analyst who provides detailed performance breakdowns for athletes. Based on the statistics provided, generate a comprehensive and insightful analysis 
    player stats:
    {json_rows[i]}
    this stats are from sofifa website so the data is very reliable
    i want the description to cover all 84 stats mentioned Above no room for error
    """
    response = generate("llama3.1:8b", prompt, options={"temperature": 0})
    response_l.append(response["response"])
    
    # Write the response data to the CSV
    with open('responses.csv', 'a', newline='', encoding='utf-8') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow([response["response"]])
```
